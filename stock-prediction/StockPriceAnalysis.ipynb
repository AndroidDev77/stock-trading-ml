```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from yahoo_fin import stock_info as si
from collections import deque
from fastai.tabular import add_datepart

import numpy as np
import pandas as pd
import random
import tensorflow as tf
import datetime

date_now = '2020-08-22'
enddatestr = datetime.datetime.strptime(date_now, "%Y-%m-%d").strftime("%m/%d/%Y")
df = si.get_data('TSLA', start_date = '01/01/2017', end_date = enddatestr, index_as_date = False)
add_datepart(df, 'date')
```

```python
list = [0, 1, 2, 3, 4]
count = pd.Series(list)
ncount = pd.Series(list)
total = len(df)/5
print(total)
print(df.index)
dfrange = range(1, df.index.stop, 1)

for idx in dfrange:
    if(df['open'].loc[idx] < df['high'].loc[idx]):
        count[df['Dayofweek'].loc[idx]] = count[df['Dayofweek'].loc[idx]] + 1
    else:
        ncount[df['Dayofweek'].loc[idx]] = ncount[df['Dayofweek'].loc[idx]] + 1
        
percent = count/total *100
npercent = ncount/total *100

print('Num Green Days ')
print(count)
print('Num Red Days ')
print(ncount)
print('Percent')
print(percent)
print('Negative Percent')
print(npercent)

print(percent + npercent)
```
